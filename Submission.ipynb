```python
topics = df['topic'].to_list()
papers = df['paper_ID'].to_list()
sent = df['pos1'].to_list()
```

```python
task = topics[0]
pid = papers[0]
l = len(topics)
sentences = []
for i in range(l):
    topic = topics[i]
    paper = papers[i]
    if(topic == task and paper == pid) or (i==0):
        sentences.append(sent[i])
    else:
        path = task + "/" + str(pid) + "/sentences.txt"
        with open(path, mode='wt', encoding='utf-8') as file:
            for x in sentences:
                file.write("%i\n" % x)
        task = topic
        pid = paper
        sentences = [sent[i]]
```
